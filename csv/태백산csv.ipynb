```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn.objects as so
import warnings
warnings.filterwarnings('ignore')

# 데이터 불러오기
merged = pd.read_csv('final_data.csv', index_col=0)
merged
```

```python
# 태백산맥 일대의 데이터만 정리
gangwon = merged.query('district_1=="강원특별자치도" and district_2 in ["정선군", "평창군", "태백시", "인제군", "양구군", "삼척시", "강릉시"]')
gangwon.to_csv('gangwon.csv')

gyeongbuk = merged.query('district_1=="경상북도" and district_2 in ["울진군", "영덕군", "영양군"]')
gyeongbuk.to_csv('gyeongbuk.csv')

taebaek_mountain = pd.concat([gangwon, gyeongbuk])
taebaek_mountain.to_csv('taebaek_mountain.csv')
```
